```python
# Create a program that:
# 1. Opens a file "names.txt" for write mode.
# 2. Writes the following names to the file, each on a new line: "Alice", "Bob", "Charlie", "Diana", "Eve".
# 3. then opens the file "names.txt" for read mode. and prints all logs
```

```python
import os
```

```python
# Step 1 & 2: Open file in write mode and write names
with open("names.txt", "w") as file:
    file.write("Alice\n")
    file.write("Bob\n")
    file.write("Charlie\n")
    file.write("Diana\n")
    file.write("Eve\n")

# Step 3: Open file in read mode and print contents
with open("names.txt", "r") as file:
    content = file.read()
    print("File contents:")
    print(content)
```

```python
with open("log.txt", "a") as f :
    f.write(" \n program execute successfully\n")

with open("log.txt", "r") as f:
    content = f.read()
    print("Log contents : \n")
    print(content)
```
